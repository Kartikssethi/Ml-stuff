```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Importing cv2_imshow for Colab

# Load image
img = cv2.imread('/content/MRI-brain-image-with-blur-and-noise-a-original-image-b-Gaussian-blur-c-Motion.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blurred = cv2.GaussianBlur(gray, (7, 7), 0)

# Apply adaptive thresholding
mask = cv2.adaptiveThreshold(blurred,
                              255,
                              cv2.ADAPTIVE_THRESH_MEAN_C,
                              cv2.THRESH_BINARY,
                              31,
                              10)

# Display results
cv2_imshow(img)
cv2_imshow(mask)
```
